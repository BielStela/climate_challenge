## Loading files

These files include data for:

+ **Tx**: maximum temperature
+ **Tn**: minimum temperature
+ **T**: mean temperature
+ **HR**: relative humidity
+ **HRN**: minimum relative humidity
+ **HRX**: maximum relative humidity
+ **DATA**: date with d/m/y h with half an hour gap
+ **PPT**: rain milimiters
+ **VVM10**: wind mean  speed at 10 height
+ **DVM10**: direction of wind at 10 height (degrees)
+ **VVX10**: maximum speed of wind at 10 height
+ **DVVX10**: maximum direction of wind at 10 height (degrees)
