# Check method for new stations

```python
from utils_reader import read_official_stations, read_real_files, official_station_adder
from utils_full import drop_function
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error
import pandas as pd
import numpy as np
from functools import reduce
```

### Dataframes

```python
OFFICIAL_ATTR = [['DATA', 'Tm', 'ESTACIO'],
                 ['DATA', 'Tm', 'ESTACIO'],
                 ['DATA', 'Tm',
                  'ESTACIO'],
                 ['DATA', 'Tm',
                  'ESTACIO']]
```

```python
def data_nx_ny():
    df = read_official_stations()
    real = read_real_files()
    merger = official_station_adder(OFFICIAL_ATTR, include_distance=False, distances=None)
    df = merger.transform(real, df)
    drop_function(df, "ESTACIO")
    drop_function(df, "DATA")
    df.drop(columns=['nx', 'ny', 'T_MIN', 'T_MAX'], inplace=True)
    return df
```

```python
df = data_nx_ny()
```

```python
def out_bag_year_error(df):
    error_dfs = []
    df['day'] = pd.to_datetime(df['day'], format='%Y-%m-%d', exact=True)
    for i in ['2012', '2013', '2014', '2015']:
        df_train = df[df['day'].dt.year.astype(str) != i]
        y = df_train['T_MEAN'].values
        X = df_train.loc[:,['ndays', 'Tm_10', 'Tm_11', 'Tm_12', 'Tm_13']].values
        scaler = StandardScaler()
        imputer = SimpleImputer()
        X_i = imputer.fit_transform(X)
        X_sc = scaler.fit_transform(X_i)
        linear = Ridge(alpha=0.1)
        linear.fit(X_sc, y)
        df_test = df[df['day'].dt.year.astype(str) == i]
        y_test = df_test['T_MEAN'].values
        X_test = df_test.loc[:,['ndays', 'Tm_10', 'Tm_11', 'Tm_12', 'Tm_13']].values
        X_t = imputer.transform(X_test)
        X_t = scaler.transform(X_t)
        y_pred = linear.predict(X_t)
        error = np.abs(y_pred - y_test)
        error_df = df_test.loc[:, ['day', 'LAT', 'LON']]
        error_df['error_' + i] = error
        error_df = error_df.groupby(['LAT', 'LON']).agg({'error_' + i: 'mean'})
        error_dfs.append(error_df)
    return error_dfs
```

```python
results = out_bag_year_error(df)
```

```python
results_merged = reduce(lambda x, y: pd.merge(x, y, on = ['LAT', 'LON']), results)
```

```python
results_merged['mean'] = results_merged.agg("mean", axis="columns")
```

```python
results_merged.reset_index(inplace=True)
```
